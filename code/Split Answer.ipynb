```python

from openai import OpenAI
api_key = 'OpenAI_api_key'
client = OpenAI(api_key=api_key)


def spilt_Answer(question, answer):

  prompt_0 = """
You will receive a pair of questions and answer, please follow the steps below to give me the answers broken down by years:
1. Split answers by year
2. Record answers in different years and under different types in a JSON format.
3. Ensure the JSON contains the keys "year", "type", and detailed parameters in answer if it have.  If no detailed classification is provided, set the "details" value to "null".
4. The JSON format should look like this:
[
  {
    "year": "2023",
    "data": [
      {
        "type": "type_here",
        "value": "value_here"
        "details": {
          "details_infomation_here": "details_value_here",
          "details_infomation_here": "details_value_here"
        }
      },
      {
        "type": "type_here",
        "value": "value_here"
        "details": {
          "details_infomation_here": "details_value_here",
          "details_infomation_here": "details_value_here"
        }
      },
      {
        "type": "type_here",
        "value": "value_here"
        "details": {
          "details_infomation_here": "details_value_here",
          "details_infomation_here": "details_value_here"
        }
      }
    ]
  },
  {
    "year": ,
    "data": [
      {
        "type": "type_here",
        "value": "value_here"
        "details": {
          "details_infomation_here": "details_value_here",
          "details_infomation_here": "details_value_here"
        }
      },
      {
        "type": "type_here",
        "value": "value_here"
        "details": {
          "details_infomation_here": "details_value_here",
          "details_infomation_here": "details_value_here"
        }
      },
      {
        "type": "type_here",
        "value": "value_here"
        "details": {
          "details_infomation_here": "details_value_here",
          "details_infomation_here": "details_value_here"
        }
      }
    ]
  }
]
5. Only output the JSON with the specified keys and in the specified order.
"""

  question_u = "Question: " + str(question)
  answer_u = "Answer: " + str(answer)
  completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
      {"role": "system", "content": prompt_0},
      {"role": "user", "content": question_u+answer_u}
    ],
    temperature=0
  )

  return completion.choices[0].message.content

```
